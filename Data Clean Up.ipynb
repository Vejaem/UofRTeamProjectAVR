```python
import pandas as pd 
import numpy as np
```

```python
unemploy_file = pd.read_csv('Data/Unemployment Data.csv')
unemploy_file.head()
```

```python
unemploy_file = unemploy_file.drop([0,1,2,3])
unemploy_file.head()
```

```python
unemploy_file = unemploy_file.drop(['Civilian non-institutional population', 'Civilian labor force', 'Unnamed: 5','Unnamed: 6', 
                                    'Unnamed: 7', 'Unnamed: 8'], axis=1)
unemploy_file.head()
```

```python
unemploy_file = unemploy_file.rename({'State and area':'State', 'Period':'Year', 'Unnamed: 2': 'Month', 
                                      'Unnamed: 9':'Unemployment Rate'}, axis=1)
unemploy_file.head()
```

```python
refined_df = pd.DataFrame(columns= ['State','Year', 'Month', 'Unemployment Rate'])

refined_df = unemploy_file[unemploy_file['Year'].astype(int) == 2019]
       
refined_df
```

```python
refined_df = refined_df.astype({'State': str, 'Year': int, 'Month': int, 'Unemployment Rate': float})

year_state_group = refined_df.groupby(['Year', 'State'])

avg_unemploy = year_state_group['Unemployment Rate'].mean()

avg_unemploy_df = pd.DataFrame(avg_unemploy)
avg_unemploy_df.head()
```

```python
avg_unemploy_df.reset_index(inplace=True)
avg_unemploy_df
```

```python
avg_unemploy_df = avg_unemploy_df.drop([18,34])
avg_unemploy_df
```

```python
poverty_df = pd.read_csv('Data/Poverty Data.csv', header=4)
poverty_df.head(10)
```

```python
poverty_df = poverty_df.drop(['Total', 'Below poverty', 'Standard error', 'Standard error.1'], axis=1)
poverty_df.head()
```

```python
poverty_df['Year']= ''
poverty_df
```

```python
filtered_df = poverty_df.iloc[:51].copy()

filtered_df['Year'] = 2019

filtered_df = filtered_df.rename(columns = {'STATE': 'State', 'Percent':'Poverty Rate', 'Year':'Year'})

filtered_df.head(52)
```

```python
merged_df = pd.merge(filtered_df, avg_unemploy_df, on='State', how='outer')

merged_df
```

```python
merged_df = merged_df.drop('Year_x', axis=1)
merged_df
```

```python
merged_df.rename(columns={'Year_y':'Year'}, inplace=True)
merged_df.head()
```

```python
merged_df = merged_df.round({'Unemployment Rate':2})
merged_df.head()
```

```python
output_file = 'Output Data/mergeddata.csv'
merged_df.to_csv(output_file)
```

```python
new_covid_df = new_covid_df.drop(['state', 'submission_date'], axis=1)
new_covid_df.rename(columns={'tot_cases':'Total Cases', 'tot_death':'Total Deaths'}, inplace=True)
new_covid_df.head()
```

```python
complete_df = merged_df = pd.merge(merged_df, new_covid_df, on='State', how='outer')
complete_df
```
